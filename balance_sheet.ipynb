## BALANCE SHEET

### EQUITY AND SHARES
1.	Treasury Shares Number: 
The number of shares that a company has repurchased and is holding in its treasury, reducing the number of shares outstanding in the   market.
2.	Ordinary Shares Number: 
The number of common shares that have been issued and are outstanding.
3.	Share Issued: 
The total number of shares that a company has issued, including both those held by investors and those held as treasury shares.
4.	Common Stock Equity: 
The equity that belongs to common shareholders, which is essentially the residual value of a company’s assets after liabilities are subtracted.
5.	Total Capitalization: 
The total value of a company’s equity and debt (i.e., the market cap plus total debt).
6.	Total Equity Gross Minority Interest: 
The total equity including the portion owned by minority shareholders in subsidiaries.
7.	Minority Interest: 
The portion of a subsidiary that is not owned by the parent company, but by outside shareholders.
8.	Stockholders Equity: 
The difference between a company’s total assets and total liabilities, representing the equity held by common shareholders.
9.	Gains Losses Not Affecting Retained Earnings: 
Changes in equity due to gains or losses that don’t directly impact retained earnings, often related to other comprehensive income.
10.	Other Equity Adjustments: 
Any other adjustments to equity that don’t fit into the main equity categories.
11.	Treasury Stock: 
The value of shares that the company has repurchased and is holding in its treasury.
12.	Retained Earnings: 
The cumulative profits that a company has retained rather than paid out as dividends.
13.	Additional Paid-In Capital: 
The amount of money investors paid above the par value of the company’s stock when purchasing shares.
14.	Capital Stock: 
The total value of all common and preferred shares issued by the company.
15.	Common Stock: 
The value of common stock issued, typically measured at the par value of the stock.

### DEBT AND LIABILITIES

1.	Net Debt: Total debt minus cash and cash equivalents. This represents the company’s actual debt load after accounting for its cash reserves.
2.	Total Debt: The sum of a company’s short-term and long-term debt obligations.
3.	Long Term Debt: Debt obligations that are due in more than one year, such as bonds or loans.
4.	Long Term Debt And Capital Lease Obligation: Includes long-term debt plus any obligations related to capital leases (leases that are considered purchases of assets).
5.	Current Debt And Capital Lease Obligation: Includes short-term debt (due within a year) plus any capital lease obligations.
6.	Current Debt: Debt that is due within one year.
7.	Other Current Borrowings: Other forms of borrowing that are due within a year, but don’t fall under traditional debt categories.
8.	Line Of Credit: Short-term borrowing facility provided by a bank or financial institution, typically with a set credit limit.
9.	Commercial Paper: A form of unsecured, short-term debt issued by a corporation, usually to finance short-term liabilities.
10.	Total Liabilities Net Minority Interest: Total liabilities after subtracting minority interest.
11.	Total Non Current Liabilities Net Minority Interest: The portion of liabilities that are due in more than one year, net of minority interests.
12.	Other Non Current Liabilities: Liabilities that don’t fall under specific categories and are due in more than one year.
13.	Employee Benefits: Liabilities related to employee benefits, such as pensions and healthcare plans.
14.	Trade and Other Payables Non Current: Non-current liabilities related to trade payables or other types of payables.
15.	Non Current Deferred Liabilities: Liabilities that have been deferred and are not due within the next year.
16.	Non Current Deferred Taxes Liabilities: Tax liabilities that are deferred to future periods and are due in more than one year.
17.	Payables And Accrued Expenses: Money owed by the company for services or goods already received, and other accrued expenses.
18.	Dividends Payable: The total amount of declared but unpaid dividends.
19.	Total Tax Payable: The total amount of taxes owed by the company but not yet paid.
20.	Income Tax Payable: Taxes owed to the government based on the company’s income, which have not yet been paid.
21.	Accounts Payable: Money that the company owes to its suppliers for goods and services purchased on credit.


### ASSETS

1.	Total Assets: The total value of everything that the company owns, including current and non-current assets.
2.	Total Non Current Assets: The value of the company’s assets that are not expected to be converted into cash within one year (e.g., property, plant, and equipment).
3.	Other Non Current Assets: Non-current assets that don’t fall into traditional categories, such as deferred tax assets or goodwill.
4.	Non Current Deferred Assets: Assets that have been deferred and are not expected to provide economic benefits within the next year.
5.	Non Current Deferred Taxes Assets: Tax assets that are deferred to future periods.
6.	Investments And Advances: Long-term investments and loans that the company has made to others.
7.	Long Term Equity Investment: The company’s investments in other companies’ equity that are expected to be held for more than one year.
8.	Investments in Subsidiaries at Cost: The amount of money invested in subsidiaries, valued at the original cost of the investment.
9.	Goodwill And Other Intangible Assets: Intangible assets such as goodwill (the excess of purchase price over the fair value of net assets in acquisitions) and intellectual property.
10.	Other Intangible Assets: Intangible assets that aren’t goodwill, such as patents, trademarks, or software.
11.	Goodwill: The excess of the purchase price paid for a company over the fair value of its net assets.
12.	Net PPE (Property, Plant, and Equipment): The value of tangible fixed assets after accounting for depreciation.
13.	Accumulated Depreciation: The total amount of depreciation that has been recorded against property, plant, and equipment since the assets were purchased.
14.	Gross PPE: The original cost of property, plant, and equipment before depreciation.
15.	Construction In Progress: The value of projects currently being built or developed, which haven’t been completed or capitalized yet.
16.	Machinery Furniture Equipment: The value of tangible assets such as machinery, furniture, and equipment.
17.	Buildings And Improvements: The value of buildings and improvements made to buildings.
18.	Land And Improvements: The value of land and improvements made to land.
19.	Properties: The total value of real estate properties owned by the company.

### CURRENT ASSETS

1.	Current Assets: Assets that are expected to be converted into cash within one year, such as inventory, accounts receivable, and cash.
2.	Other Current Assets: Miscellaneous current assets that don’t fit into other categories.
3.	Inventory: The value of goods available for sale that have not yet been sold.
4.	Finished Goods: The value of completed goods that are ready for sale.
5.	Raw Materials: The value of raw materials that the company has on hand for future production.
6.	Receivables: Money owed to the company by customers for goods or services already delivered.
7.	Other Receivables: Other forms of receivables, such as loans or advances made to employees or other parties.
8.	Accounts Receivable: The total amount owed to the company by customers for goods or services provided on credit.
9.	Allowance For Doubtful Accounts Receivable: The estimated amount of accounts receivable that the company expects will not be collected.
10.	Gross Accounts Receivable: The total amount of accounts receivable before considering any allowance for doubtful accounts.
11.	Cash Cash Equivalents And Short Term Investments: The total value of cash, short-term investments, and other highly liquid assets.
12.	Cash And Cash Equivalents: The value of cash on hand and other highly liquid investments that can be converted into cash quickly.

### Relevant Items for Intrinsic Value Calculation:

#### Net Debt:
   
    •	Definition: Total debt minus cash and cash equivalents.
    •	Importance: Reflects the company’s actual debt load after accounting for available cash, which is crucial for understanding the company’s financial leverage and valuation.

#### Total Debt:
   
    •	Definition: The sum of long-term debt and short-term (current) debt.
    •	Importance: Helps in assessing the company’s total liabilities and financial risk.
	
#### Tangible Book Value:
   
    •	Definition: Book value of the company’s assets minus intangible assets (like goodwill) and liabilities.
    •	Importance: Provides a measure of the company’s net worth based on tangible assets, useful in valuing the company.
	
#### Invested Capital:

    •	Definition: Total amount of capital invested in the company, including equity and debt.
    •	Importance: Indicates how much capital has been deployed into the business and helps in assessing return on invested capital (ROIC).
	
#### Working Capital:

    •	Definition: Current assets minus current liabilities.
    •	Importance: Measures the company’s short-term financial health and operational efficiency.

#### Net Tangible Assets:

    •	Definition: Total assets minus intangible assets and liabilities.
    •	Importance: Provides insight into the value of tangible assets available to shareholders.

#### Common Stock Equity:

    •	Definition: Equity attributable to common shareholders, often equal to total equity minus preferred equity.
    •	Importance: Represents the ownership interest of common shareholders, crucial for equity valuation.

#### Total Capitalization:

    •	Definition: Sum of total debt and total equity.
    •	Importance: Reflects the total value of the company’s capital structure, used in various valuation metrics.

#### Stockholders Equity:

    •	Definition: Total assets minus total liabilities.
    •	Importance: Represents the residual interest in the assets of the company after deducting liabilities.

#### Retained Earnings:

    •	Definition: Accumulated net income that has not been distributed to shareholders as dividends.
    •	Importance: Reflects the portion of profits reinvested in the company, important for assessing growth potential.

#### Total Liabilities:

    •	Definition: The sum of all short-term and long-term liabilities.
    •	Importance: Provides a complete picture of the company’s obligations and financial risk.

```python
# Define the list of debt-related items
debt_items = [
    'Long Term Debt',
    'Current Debt',
    'Cash And Cash Equivalents',
    'Net Debt',
    'Total Debt'
    
]

# Define the list of other financial metrics
other_items1 = [
    'Tangible Book Value',
    'Net Tangible Assets',
    'Retained Earnings'
]

other_items2 = [
    
    'Invested Capital',
    'Common Stock Equity',
    'Total Capitalization',
    'Stockholders Equity',
]

# Extract and convert the data for debt-related items
debt_items_df = balance_sheet.loc[debt_items] / 1_000_000  # Convert to millions of USD

# Extract and convert the data for other financial metrics
other_items1_df = balance_sheet.loc[other_items1] / 1_000_000  # Convert to millions of USD
other_items2_df = balance_sheet.loc[other_items2] / 1_000_000  # Convert to millions of USD
# Transpose and clean data for both tables
debt_items_clean_df = debt_items_df.T.dropna()
other_items1_clean_df = other_items1_df.T.dropna()
other_items2_clean_df = other_items2_df.T.dropna()

# Convert index to year only
debt_items_clean_df.index = pd.to_datetime(debt_items_clean_df.index).year
other_items1_clean_df.index = pd.to_datetime(other_items1_clean_df.index).year
other_items2_clean_df.index = pd.to_datetime(other_items2_clean_df.index).year

# Convert to DataFrames for tabulate
debt_df = pd.DataFrame(debt_items_clean_df)
other1_df = pd.DataFrame(other_items1_clean_df)
other2_df = pd.DataFrame(other_items2_clean_df)

# Print the tables in a readable format
print("Debt-Related Items Table:")
print(tabulate(debt_df, headers='keys', tablefmt='grid', floatfmt='.2f'))

print("\nOther Financial Metrics Table 1:")
print(tabulate(other1_df, headers='keys', tablefmt='grid', floatfmt='.2f'))

print("\nOther Financial Metrics Table 2:")
print(tabulate(other2_df, headers='keys', tablefmt='grid', floatfmt='.2f'))
```

```python
shareholders_equity = balance_sheet.loc['Stockholders Equity'] / 1_000_000  # Convert to MUSD
```

```python
total_assets = balance_sheet.loc['Total Assets'] / 1_000_000  # Convert to MUSD
```

```python
invested_capital = balance_sheet.loc['Invested Capital'] / 1_000_000  # Convert to MUSD
```

```python
# Calculate financial ratios
roe = (net_income / shareholders_equity) * 100  # Return on Equity
roa = (net_income / total_assets) * 100  # Return on Assets
roi = (net_income / invested_capital) * 100  # Return on Investment

# Combine the results into a DataFrame for easier reading
ratios_df = pd.DataFrame({
    'ROE (%)': roe,
    'ROA (%)': roa,
    'ROI (%)': roi
})

fin_info_df = pd.DataFrame({
    'Net Income (MUSD)': net_income,
    'Shareholders Equity (MUSD)': shareholders_equity,
    'Total Assets (MUSD)': total_assets,
    'Invested Capital (MUSD)': invested_capital
})
```

```python
# Clean the DataFrame to drop rows with any NaN values
ratios_df = ratios_df.dropna()
fin_info_df = fin_info_df.dropna()
print("\nFinancial Ratios by Year:")
print(tabulate(ratios_df, headers='keys', tablefmt='grid', floatfmt='.2f'))
```
